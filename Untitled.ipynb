## ClÃºsteres (Heber)

```python
vdf3= vdf1
```

### DEJAMOS SOLO LAS COLUMNAS QUE NO PRESENTAN UNA ALTA CORRELACION ENTRE ELLAS COMO SER UNDER-FIVE-DEATHS CON INFANT-DEATHS

```python
columnas_a_usar = [ 'Life expectancy ', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'Hepatitis B',
       'Measles ', ' BMI ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', 'Schooling']
```

```python
vdfx  = vdf3.loc[ : , columnas_a_usar].values
```

### Estandarizamos los valores

```python
StandardScaler().fit_transform(vdfx)
```

### Reducimos a dos componentes haciendo uso de PCA con el objetivo de que se concentre la informacion en las dos primeras componentes 

```python
pcavdf3 = PCA(n_components=2)
```

```python
componentes_principales = pcavdf3.fit_transform(vdfx)
```

```python
pcadfv = pd.DataFrame(data=componentes_principales , columns=['c1','c2'])
```

```python
kmeans = KMeans(n_clusters=2).fit(pcadfv)
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
plt.scatter(pcadfv['c1'], pcadfv['c2'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```python
final_vdf = vdf3.loc[ : , columnas_a_usar ]
```

```python
final_vdf["cluster"] = kmeans.labels_
final_vdf.cluster = final_vdf.cluster.astype(str)
```

```python
final_vdf
```

```python
loadings = pcavdf3.components_.T * np.sqrt(pcavdf3.explained_variance_)
```

```python
fig = px.scatter(componentes_principales, x=0, y=1, color=final_vdf['cluster'])
```

```python
for i, feature in enumerate(columnas_a_usar):
    fig.add_shape(
        type='line',
        x0=0, y0=0,
        x1=loadings[i, 0],
        y1=loadings[i, 1]
    )
    fig.add_annotation(
        x=loadings[i, 0],
        y=loadings[i, 1],
        ax=0, ay=0,
        xanchor="center",
        yanchor="bottom",
        text=feature,
    )
fig.show()
```

```python
final_vdf_cluster_0 = final_vdf.loc[ final_vdf['cluster'] == '0'  , : ]
final_vdf_cluster_1 = final_vdf.loc[ final_vdf['cluster'] == '1'  , : ]
```
