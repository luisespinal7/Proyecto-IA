## Análisis de Calidad de los Datos

### Análisis de Valores Nulos
Porcentaje de valores nulos por cada columna del dataset, realizando un análisis por anio.

```python
nan_analysis = pd.DataFrame()
```

```python
for Year_ in life_expectancy.loc[ : , "Year" ].drop_duplicates():
    countrydf = life_expectancy.loc[ life_expectancy['Year'] == Year_ , : ]    
    for column_ in countrydf.columns:        
        percentage =  np.sum(pd.isna( countrydf.loc[ : , column_ ]  )) / countrydf.shape[0]
        dfrow = pd.DataFrame({
            'Year': [ Year_ ]
            , 'column' : [ column_ ]
            , 'percentage.of.nans' : [ percentage ]
        })
        nan_analysis = pd.concat([nan_analysis, dfrow])
```

```python
# Cálculo del número de valores atípicos y su respectivo porcentaje utilizando el método Tukey.

for variable in boxplots_columnas.keys():
    q75, q25 = np.percentile(vdf1[variable], [75 ,25])
    iqr = q75 - q25

    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    print("El número de valores atípicos para {} es {} y su porcentaje es : {}".format(variable,
                                                                             len((np.where((vdf1[variable] > max_val) | 
                                                                                           (vdf1[variable] < min_val))[0])),len((np.where((vdf1[variable] > max_val) | 
                                                                                           (vdf1[variable] < min_val))[0]))*100/1987))
```

Para remover los valores atípicos y mostrar los boxplots sin ellos utilizando la técnica Winsorize:
